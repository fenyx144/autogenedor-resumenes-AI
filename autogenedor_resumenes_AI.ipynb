```python
from transformers import pipeline

# Cargar el pipeline de Hugging Face para resúmenes
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Función para resumir un texto
def generar_resumen(texto):
    resumen = summarizer(texto, max_length=150, min_length=50, do_sample=False)
    return resumen[0]['summary_text']

# Ejemplo de texto
texto_largo = """
    La inteligencia artificial (IA) es un campo de estudio que se ocupa de crear sistemas que imitan la inteligencia humana.
    La IA puede aprender y tomar decisiones por sí misma. Esta tecnología está presente en múltiples áreas como la
    medicina, la ingeniería, la robótica, y la atención al cliente. La IA puede ser clasificada en dos categorías principales:
    IA débil e IA fuerte. La IA débil está diseñada para realizar tareas específicas, como los asistentes virtuales,
    mientras que la IA fuerte tiene la capacidad de realizar cualquier tarea cognitiva humana. A pesar de sus enormes
    beneficios, la IA también plantea desafíos éticos y laborales que deben ser abordados.
"""

# Generar el resumen
resumen_generado = generar_resumen(texto_largo)

# Imprimir el resumen
print("Resumen Generado:")
print(resumen_generado)
```
